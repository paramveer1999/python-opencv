```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
#IMPORTING THE DATASET
img=cv2.imread("DATA/crossword.jpg",0)
```

```python
#CONVERTING THE GBR TO RGB FORMAT
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
#INCREASING THE SIZE OF IMAGE
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(111)
ax.imshow(img1,cmap="gray")
plt.title("ORIGINAL IMAGE")
```

```python
#APPLYING DIFFERENT THRESHOLDING ALGORITHMS ON ABOVE IMAGE
```

## BINARY

```python
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(111)
#ax.imshow(img1,cmap="gray")
ret,o2=cv2.threshold(img1,127,255,cv2.THRESH_BINARY)
plt.imshow(o2)
plt.title("BINARY")
```

## BINARYINVERSE

```python
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(111)
#ax.imshow(img1,cmap="gray")
ret,o2=cv2.threshold(img1,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(o2)
plt.title("BINARY_INVERSE")
```

## ZERO

```python
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(111)
#ax.imshow(img1,cmap="gray")
ret,o2=cv2.threshold(img1,127,255,cv2.THRESH_TOZERO)
plt.imshow(o2)
plt.title("ZERO")
```

## ZERO_INVERSE

```python
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(111)
##ax.imshow(img1,cmap="gray")
ret,o2=cv2.threshold(img1,127,255,cv2.THRESH_TOZERO_INV)
plt.imshow(o2)
plt.title("ZERO_INVERSE")
```
