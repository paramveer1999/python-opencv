## Morphological gradient helps more in detecting edge of images

```python
#CREATING TWO FUNCTIONS LOAD_IMG() AND DIAPLY_IMG()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def load_img():
    img=cv2.imread("DATA/sudoku.jpg",0)
    result=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return result
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
i=load_img()
display_img(i)
```

```python
#WE HAVE SOBELX AND SOBELY OPERATOR IN THIS
#SOBELX----->IT HELPS IN DETERMINING THE VERTICAL EDGES
#SOBELY------>IT HELPS IN DETERMINING THE HORIZONTAL EDGES
```

```python
sobelx=cv2.Sobel(src=i,ddepth=cv2.CV_64F,dx=1,dy=0,ksize=3)
display_img(sobelx)
```

```python
##SOBEL Y
```

```python
result=cv2.Sobel(src=i,ddepth=cv2.CV_64F,dx=0,dy=1,ksize=3)
display_img(result)
```

```python
#LAPLACIAN -->>it will automatically calculate sobelx and sobely
```

```python
laplacian=cv2.Laplacian(i,cv2.CV_64F)
display_img(laplacian)
```
