```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("DATA/sammy_face.jpg",0)
```

```python
plt.imshow(img,cmap="gray")
```

```python
img1=cv2.Canny(img,10,97)#L2gradient=false-->g=[g1]+[g2]
plt.imshow(img1,cmap="gray")
```

```python
img1=cv2.Canny(img,10,97,L2gradient=True)##g=root(g1*2+g2*2)
plt.imshow(img1,cmap="gray")
```

```python
img=cv2.blur(img,ksize=(5,5))
plt.imshow(img)
```

## usually in the canny method kernel size of 3*3 is used if we want ot blur the image by our own kernel size then we can use "CV2.BLUR()" method

```python
img1=cv2.Canny(img,10,97,L2gradient=True)
plt.imshow(img1,cmap="gray")
```
