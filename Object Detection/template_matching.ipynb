```python
methods=["cv2.TM_CCOEFF","cv2.TM_CCOEFF_NORMED","cv2.TM_CCORR","cv2.TM_CCORR_NORMED","cv2.TM_SQDIFF","cv2.TM_SQDIFF_NORMED"]
```

```python
methods=["cv2.TM_CCOEFF","cv2.TM_CCOEFF_NORMED","cv2.TM_CCORR","cv2.TM_CCORR_NORMED","cv2.TM_SQDIFF","cv2.TM_SQDIFF_NORMED"]
```

```python
for i in methods:
    
    #creating the co[py of tha image
    full_copy=face.copy()
    
    method=eval(i)
    
    #TEMPLATE MATCHING
    res=cv2.matchTemplate(full_copy,big_face,method)
    
    
    #GRABING THE LOCATION FOR rectangle
    
    minVal, maxVal, minLoc, maxLoc=cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left=minLoc
    else:
        top_left=maxLoc
    
    ##BOTTOM RIGHT LOCATION
    height,width,channel=big_face.shape
    bottom_right=(top_left[0]+width,top_left[1]+height)
    cv2.rectangle(full_copy,pt1=top_left,pt2=bottom_right,color=(255,0,0),thickness=10)
    
    #plot and show image
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title("HEAT MAP")
    
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title("DETECTION OF TEMPLATE")
    plt.suptitle(i)
    plt.show()
    #
    #plt.suptitle(i)
```
