```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#LOADING THE IMAGES
img1=cv2.imread("DATA/the_book_thief.JPG",0)
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.subplot(121)
print(plt.imshow(img1))
img2=cv2.imread("DATA/me_holding_book.JPG",0)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.subplot(122)
plt.imshow(img2)
```

```python
#creating a orb detector
```

```python
orb=cv2.ORB_create()
```

```python
keypoint1,des1=orb.detectAndCompute(img1,None)
keypoint2,des2=orb.detectAndCompute(img2,None)
```

```python
#BRUTE FORCE MATCHING
```

```python
bf=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

```python
matches=bf.match(des1,des2,None)
```

```python
for m in matches:
    print(m.distance)
```

```python
#Sorting the matches based on descriptor values
```

```python
matches=sorted(matches,key=lambda x:x.distance)
```

```python
for m in matches:
    print(m.distance)
```

```python
##BASED ON OUR DESCRIPTOR MATCHING WE WIIL COMPUTE THE RESULT
```

```python
matching_result=cv2.drawMatches(img1,keypoint1,img2,keypoint2,matches,None)
```

```python
plt.imshow(matching_result)
```

```python
#TAKING ONLY FIRST 41 DESCRIPTOR VALUES
```

```python
matching_result=cv2.drawMatches(img1,keypoint1,img2,keypoint2,matches[:41],None)
```

```python
plt.imshow(matching_result)
```
