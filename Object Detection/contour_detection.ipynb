```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1=cv2.imread("DATA/00-puppy.jpg")
img2=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
```

```python
img3=cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
plt.imshow(img3)
```

```python
#APPLYING BINARY THRESHOLD TO THE IMAGE
```

```python
ret,thresh=cv2.threshold(img3,127,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh)
```

```python
contours,hie=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
```

```python
cv2.drawContours(img2,contours,5,(0,255,0),10)
plt.imshow(img2)
```
